Machine Learning Intro:

Machine Learning (ML) is about teaching computers to learn
patterns from data and make predictions or decisions without 
being explicitly  progrmmed.

Supervised Learning:
    Data has labels (Correct Answers)

    Example: Training with fruit images
    labelled "Apple" or "Banana".

    Used for regression (Predict Numbers)
    and Classification (Predict Category)

Unsupervised Learning:

    Data has no labels

    The model groups or fiinds hidden patterns

    Example: Grouping customers by shopping habbits without 
    knowing categories.
