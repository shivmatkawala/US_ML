```python
# K-Means is an unsupervised machine learning algorithm
# that groups similar data into k clusters by repetedly assigning
# each point to the nearest centroid and updating the centroid
# to the average of its cluster until the cluster stops changing.
```

```python
# Imports:
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("dataFiles/wholesale_customers_data.csv")
df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Check Missing Values
df.isnull().sum()
```

```python
X = df.drop(["Channel", "Region"], axis=1)
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[:5]
```

```python
# Find Best K: (Elbow Method)

wcss = []

for k in range(1, 11):
    model = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=10
    )

    model.fit(X_scaled)

    wcss.append(model.inertia_)
wcss
```

```python
plt.figure(figsize=(8,5))

plt.plot(range(1, 11), wcss, marker="o")
plt.xlabel("Number of clusters (K)")
plt.ylabel(("WCSS"))

plt.title("Elbow Method")

plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)


kmeans.fit(X_scaled)
```

```python
clusters = kmeans.predict(X_scaled)
clusters
```

```python
df["Cluster"] = clusters

df.head()
```

```python
print(kmeans.cluster_centers_)
```

```python
print(df["Cluster"].value_counts())
```

```python
score = silhouette_score(
    X_scaled,
    clusters
)

print(score)
```

```python
# Visualization using PCA:

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(9, 6))
plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=clusters,
    cmap="viridis",
    s=60
)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.title("Customer Clusters")

plt.colorbar(label="Cluster")

plt.show()
```

```python
centers = pca.transform(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(9, 6))
plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    c=clusters,
    cmap="viridis",
    s=60
)

plt.scatter(
    centers[:,0],
    centers[:,1],
    c="red",
    marker="X",
    s=300,
    label="Centroids"
)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")

plt.title("Clusters with Cetroids")
plt.legend()
plt.show()
```
